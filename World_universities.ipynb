## Data Preparation and Cleaning

World universities ranking data is expressed in pandas dataframe.Out of 100 each university is given a score
for their performance in that concerned vertical.
for example teaching resources, income, research, international outreach.etc and finally they are ranked accordingly to their total performance score.
As I reside in india,I extracted data of universities in my country india.I am modifying some data in it.

```python
import pandas as pd
```

```python
world_univ_rankings_df = pd.read_csv('./timesData.csv')
```

```python
world_univ_rankings_df
```

```python
world_univ_rankings_df.info()
```

```python
world_univ_rankings_df.describe()
```

```python
ranking_2016_df = world_univ_rankings_df[world_univ_rankings_df.year == 2016]
ranking_2016_df
```

```python
indian_ranking2016_df = ranking_2016_df[ranking_2016_df.country == 'India']
indian_ranking2016_df
```

```python
##As total score is not mentioned i am removing that column and adding a new column for total score that's based on my formula.
indian_ranking2016_df.drop(columns = ['total_score'], inplace=True)
```

```python
## as indian universities total scores are missing, i am changing the method of calculation of total score 
indian_ranking2016_df['total_score'] = (indian_ranking2016_df.teaching + indian_ranking2016_df.research + indian_ranking2016_df.citations)/3
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

As the actual data is very large,so data of USA's universities rankings for the year of 2016 is analysed.  

```python
USA_ranking_df = ranking_2016_df[ranking_2016_df.country == 'United States of America']
USA_ranking_df
```

A scatterplot is plotted between 'research' versus 'citations'.

As it can be observed a good number of educational institutes scored good in 'citations(RESEARCH INFLUENCE)'. as well as in 'research'.

```python
plt.figure(figsize=(30, 30))
sns.scatterplot('research', 'citations', hue='university_name', data=USA_ranking_df, s=500);
```
